```python
from sys import maxsize
from itertools import permutations
V = 6    
cost=[]
pa=[]
def travellingSalesmanProblem(graph, s):
	vertex = []
	for i in range(V):
		if i != s:
			vertex.append(i)
	min_path = maxsize
	next_permutation=permutations(vertex)
	for i in next_permutation:
		current_pathweight = 0
		k = s
		for j in i:
			current_pathweight += graph[k][j]
			k = j
		current_pathweight += graph[k][s]
		pa.append(list(i))

		min_path = min(min_path, current_pathweight)
		cost.append(min_path)
		
	return min_path
if __name__ == "__main__":
	graph = [[0, 10, 15, 20, 13, 8], [10, 0, 35, 25, 19, 28],        
			[15, 35, 0, 30, 18, 26], [20, 25, 30, 0, 16, 22],[13, 19, 18, 16, 0, 11],[8, 28, 26, 22, 11, 0]]
	s = 0
	
	sale=travellingSalesmanProblem(graph, s)
	index=cost.index(sale)
	print("The minimum cost path starting with 0 is",*pa[index],"with cost",sale)
	print("OTHER POSSIBLE PATHS WITH COST:")
	for i in range(len(cost)):
	    print("path no",i,":",pa[i]," cost:",cost[i])
	    print("\n")
```
